```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages

data = pd.read_csv('data.csv')

years = data['Year'].unique()
universities = data['University'].unique()
majors = data['Major'].unique()
times = sorted(data['Time'].unique())


with PdfPages("Charts/Distributions.pdf") as out_fileD:
    plt.figure(figsize=(8, 12))
    sns.countplot(x='Major', data=data, order=data['Major'].value_counts().index)
    plt.xticks(rotation=90)
    plt.xlabel("Major")
    plt.ylabel("Count")
    plt.title("Distribution of Majors")
    plt.tight_layout()
    out_fileD.savefig(plt.gcf())
    plt.close()
    
    plt.figure(figsize=(8, 12))
    sns.countplot(x='University', data=data, order=data['University'].value_counts().index)
    plt.xticks(rotation=90)
    plt.xlabel("University")
    plt.ylabel("Count")
    plt.title("Distribution of Universities")
    plt.tight_layout()
    out_fileD.savefig(plt.gcf())
    plt.close()
    
    plt.figure(figsize=(8, 12))
    sns.countplot(x='Order', data=data, order=data['Order'].value_counts().index)
    plt.xticks(rotation=90)
    plt.xlabel("Order")
    plt.ylabel("Count")
    plt.title("Distribution of Orders")
    plt.tight_layout()
    out_fileD.savefig(plt.gcf())
    plt.close()
    
    plt.figure(figsize=(8, 12))
    sns.countplot(x='Year', data=data, order=data['Year'].value_counts().index)
    plt.xticks(rotation=90)
    plt.xlabel("Year")
    plt.ylabel("Count")
    plt.title("Distribution of Years")
    plt.tight_layout()
    out_fileD.savefig(plt.gcf())
    plt.close()

with PdfPages("Charts/Orders_By_Year.pdf") as out_file:
    fig, axs = plt.subplots(len(years), 1, figsize=(12, 8 * len(years)))
    for i, year in enumerate(years):
        subset = data[data['Year'] == year]
        sns.countplot(x='Order', data=subset, ax=axs[i])
        axs[i].set_title(f'Order Count for {year}')
        axs[i].set_xlabel('Order')
        axs[i].set_ylabel('Count')
        axs[i].tick_params(axis='x', rotation=90)
    plt.tight_layout()
    out_file.savefig(plt.gcf())

with PdfPages("Charts/Orders_By_University.pdf") as out_file2:
    fig, axs = plt.subplots(len(universities), 1, figsize=(10, 6 * len(universities)))
    for i, university in enumerate(universities):
        subset = data[data['University'] == university]
        sns.countplot(x='Order', data=subset, ax=axs[i])
        axs[i].set_title(f'Order Count for University {university}')
        axs[i].set_xlabel('Order')
        axs[i].set_ylabel('Count')
        axs[i].tick_params(axis='x', rotation=90)
    plt.tight_layout()
    out_file2.savefig(plt.gcf())

with PdfPages("Charts/Orders_By_Major.pdf") as out_file3:
    fig, axs = plt.subplots(len(majors), 1, figsize=(10, 6 * len(majors)))
    for i, major in enumerate(majors):
        subset = data[data['Major'] == major]
        sns.countplot(x='Order', data=subset, ax=axs[i])
        axs[i].set_title(f'Order Count for Major {major}')
        axs[i].set_xlabel('Order')
        axs[i].set_ylabel('Count')
        axs[i].tick_params(axis='x', rotation=90)
    plt.tight_layout()
    out_file3.savefig(plt.gcf())

with PdfPages("Charts/Orders_By_Time.pdf") as out_file4:
    
    fig, axs = plt.subplots(len(times), 1, figsize=(10, 6 * len(times)))
    for i, major in enumerate(times):
        subset = data[data['Time'] == major]
        sns.countplot(x='Order', data=subset, ax=axs[i])
        axs[i].set_title(f'Order Count for Time {major}')
        axs[i].set_xlabel('Order')
        axs[i].set_ylabel('Count')
        axs[i].tick_params(axis='x', rotation=90)
    plt.tight_layout()
    out_file4.savefig(plt.gcf())
```
